# Calculating Weighted Average for movie's rating
### W= (R*v+C*m)/(v+m)
#### R - avg rating for the movie
#### v - number of votes for the movie
#### m - minimum votes reqd. to be listed in top 250
#### C - mean vote across the whole report

```python
# Calculating the parameters for W
v= movies_cleaned['vote_count']
R = movies_cleaned['vote_average']
C = movies_cleaned['vote_average'].mean()
m = movies_cleaned['vote_count'].quantile(0.7)
```

```python
movies_cleaned['weighted_average'] = ((R*v+C*m)/(v+m))
```

```python
movies_sorted_ranking = movies_cleaned.sort_values('weighted_average', ascending=False)
movies_sorted_ranking[['original_title', 'vote_count', 'vote_average', 'weighted_average',
                       'popularity']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
weighted_average = movies_sorted_ranking.sort_values('weighted_average', ascending=False)
plt.figure(figsize =(12,6))
axis1 = sns.barplot(x=weighted_average['weighted_average'].head(10),
                    y=weighted_average['original_title'].head(10))
plt.title('Best Movies by avg votes')
plt.xlabel('Avg. Score')
plt.ylabel('Movie Title')
plt.show()
```

```python
popularity = movies_sorted_ranking.sort_values('popularity', ascending=False)
plt.figure(figsize =(12,6))
axis1 = sns.barplot(x=popularity['popularity'].head(10),
                    y=popularity['original_title'].head(10))
plt.title('Best Movies by avg votes')
plt.xlabel('Popularity Score')
plt.ylabel('Movie Title')
plt.show()
```

## Recommendation based on scaled weight

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
movie_scaled = scaler.fit_transform(movies_cleaned[['weighted_average', 'popularity']])
movie_normalized = pd.DataFrame(movie_scaled, columns=['weighted_average', 'popularity'])
movie_normalized.head()
```

```python
movies_cleaned[['normalized_weighted_average', 'normalized popularity']] = movie_normalized
```

```python
movies_cleaned.head()
```

```python
movies_cleaned['score'] = movies_cleaned['normalized_weighted_average']*0.5+movies_cleaned['normalized popularity']*0.5

movies_score = movies_cleaned.sort_values(['score'], ascending=False)
movies_score[['original_title', 'normalized_weighted_average', 'normalized popularity', 'score']].head()
```
